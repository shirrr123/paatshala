```python
#1.a. Display the Glucose level and the age for samples having the index ranging from 45 to 60
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('diabetes_model.csv')
glucose_age_subset=df.loc[45:60,['Glucose','Age']]
print("\nGlucose  level and age for samples 45 to 60:")
print(glucose_age_subset)
print(df)
```

```python
#2b. How many patients in the data are having the blood pressure level above 120?
bp_count=df[df['BloodPressure']>120].shape[0]
print("\nNumber of patients with blood pressure level above 120:",bp_count)
```

```python
#3c. Find the minimum and maximum value of insulin level in the data
min_insulin=df['Insulin'].min()
max_insulin=df['Insulin'].max()
print("\nMinimum insulin level:",min_insulin)
print("\nMaximum insulin level:",max_insulin)
```

```python
#4d. Create a histogram of "Age"
plt.hist(df['Age'],bins=10,edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
#5 e. Create a scatter plot of "BMI" vs "Glucose
plt.scatter(df['BMI'],df['Glucose'])
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.title('Scatter Plot of BMI vs Glucose')
plt.show()
```
